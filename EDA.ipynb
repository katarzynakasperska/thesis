```python
import pandas as pd
data = pd.read_excel('/content/drive/MyDrive/thesis/subset.xlsx')
data.isnull().sum()
```

```python
#see distribution of lengths of the songs, sometimes the program might have captured non-lyrics but treated them as lyrics
import seaborn as sns
length = data['length'].tolist()
sns.boxplot(x=length)
```

```python
rhyme_score = data['rhyme_score'].tolist()
sns.boxplot(x=rhyme_score)
```

```python
#correlation all rhymes

hit_list = data['hit'].tolist()
score_list = data['rhyme_score'].tolist()

from scipy import stats
res = stats.pearsonr(hit_list, score_list)
res
```

```python
#correlation alternating rhymes

hit_list = data['hit'].tolist()
score_list = data['alternating'].tolist()

from scipy import stats
res = stats.pearsonr(hit_list, score_list)
res
```

```python
#correlation coupled rhymes

hit_list = data['hit'].tolist()
score_list = data['coupled'].tolist()

from scipy import stats
res = stats.pearsonr(hit_list, score_list)
res
```

```python
#correlation same words

hit_list = data['hit'].tolist()
score_list = data['same'].tolist()

from scipy import stats
res = stats.pearsonr(hit_list, score_list)
res
```

```python
#correlation second-person pronouns

hit_list = data['hit'].tolist()
score_list = data['norm_you'].tolist()

from scipy import stats
res = stats.pearsonr(hit_list, score_list)
res
```
